```python
import numpy as np
from scipy.stats import pearsonr
x = np.arange(1, 6)
y = 2*x

corr, p = pearsonr(x, y)
print(corr)
```

```python
y = [2, 40, 67, 800, 111230]
corr, p = pearsonr(x, y)
print(corr)
```

```python
y = np.random.rand(5)
corr, p = pearsonr(x, y)
print(corr)
```

```python
x = np.random.normal(0, 1, 100)
y = x + np.random.normal(0, .1, 100)

r, p = pearsonr(x, y)

print(f"PCC: {r}; p-val: {p}; Significant: {p < 0.05}")
```

```python
x = np.arange(1, 6)
y = 2*x

import matplotlib.pyplot as plt

plt.scatter(x,y)
```

```python
np.random.seed(182)
x = np.random.normal(0, 1, 100)
y = x + np.random.normal(0, .1, 100)

plt.scatter(x,y)
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x, p(x), 'r')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Title');
```
